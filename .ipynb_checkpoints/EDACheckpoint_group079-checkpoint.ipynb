```python
#initialize all the data
df_2007 = pd.read_sas('PAQ_E.XPT')
df_2009 = pd.read_sas('PAQ_F.XPT')
df_2011 = pd.read_sas('PAQ_G.XPT')
df_2013 = pd.read_sas('PAQ_H.XPT')
df_2015 = pd.read_sas('PAQ_I.XPT')
```

## Data Cleaning
Describe your data cleaning steps here

*Code mainly done by Nick

This cell contains all of the general functions that would then be applied to each year’s dataset. We created a function to fix very small numbers within the datasets. Some of the numbers were very close to zero but not quite, so we needed a way to fix these “off” numbers. Next we made a function in order to standardize the times. Some of the datasets used minutes and some used hours. In order to not lose the smaller details of the minute data, we made a function to convert the hours into minutes. We also made a function to correctly assess some of the weird entries given by the data set about computer usage (i.e. 99 meaning does not apply). Finally we created a function to correctly assess the weird entries given by activities (ie having very large numbers that couldn’t possibly be real).


```python
#clean data functions
def fix_small_numbers(row): #turn numbers less than 0 to 0
    if np.isnan(row) == True:
        return row
    else:
        return int(row)
def standardize_time(row): #minutes to hours
    if np.isnan(row) == True:
        return row
    else:
        ans = row * 60
        return ans 
def clean_computer_TV(row):
    if np.isnan(row) == True:
        return row
    if int(row) == 8:
        return None
    if int(row) == 6:
        return None
    if int(row) == 77:
        return None
    if int(row) == 99:
        return None
    else:
        return row

def clean_activities(row):
    if np.isnan(row) == True:
        return row
    if int(row) == 7777:
        return None
    if int(row) == 9999:
        return None
    if int(row) > 480:
        return None
    else:
        return row
```

To clean our data, we dropped columns in our data sets from the years 2007 to 2015 that we did not need. We dropped these columns because they contained generalized and unrelated  information that  had no clear relationship with the variables we are analyzing. After removing the unnecessary columns, we used our data cleaning functions we created above (applied to the specific year datasets that needed them) in order to further clean the four columns that we planned to use

```python
#clean data
def clean_2007(df):
    df = df.rename(columns = {'SEQN': 'User ID','PAD590':'minutes watch TV or videos past 30 days','PAD600': '# of minutes use computer past 30 days','PAD660':'minutes vigorous recreational activities','PAD675':'minutes moderate recreational activities'}).drop(columns=['PAD615','PAD630','PAD645','PAD680','PAQ560','PAQ605','PAQ610','PAQ620','PAQ625','PAQ635','PAQ640','PAQ650','PAQ655','PAQ665','PAQ670','PAAQUEX' ])
    tv_comp = ['minutes watch TV or videos past 30 days','# of minutes use computer past 30 days' ]
    recreation = ['minutes vigorous recreational activities','minutes moderate recreational activities']
    for i in tv_comp:
        df[i] = df[i].apply(clean_computer_TV)
    for i in recreation:
        df[i] = df[i].apply(clean_activities)
    for i in df.columns:
        df[i] = df[i].apply(fix_small_numbers)
    for i in tv_comp:
        df[i] = df[i].apply(standardize_time)
    return df
def clean_2009(df):
    df= df.rename(columns = {'SEQN': 'User ID','PAD590':'minutes watch TV or videos past 30 days','PAD600': '# of minutes use computer past 30 days','PAD660':'minutes vigorous recreational activities','PAD675':'minutes moderate recreational activities'}).drop(columns=['PAQ706','PAD615','PAD630','PAD645','PAD680','PAQ605','PAQ610','PAQ620','PAQ625','PAQ635','PAQ640','PAQ650','PAQ655','PAQ665','PAQ670','PAAQUEX' ])
    tv_comp = ['minutes watch TV or videos past 30 days','# of minutes use computer past 30 days' ]
    recreation = ['minutes vigorous recreational activities','minutes moderate recreational activities']
    for i in tv_comp:
        df[i] = df[i].apply(clean_computer_TV)
    for i in recreation:
        df[i] = df[i].apply(clean_activities)
    for i in df.columns:
        df[i] = df[i].apply(fix_small_numbers)
    for i in tv_comp:
        df[i] = df[i].apply(standardize_time)
    return df
def clean_2011(df):
    df = df.rename(columns = {'SEQN': 'User ID','PAQ710':'minutes watch TV or videos past 30 days','PAQ715': '# of minutes use computer past 30 days','PAD660':'minutes vigorous recreational activities','PAD675':'minutes moderate recreational activities'}).drop(columns=['PAQ706','PAD615','PAD680','PAD630','PAD645','PAQ605','PAQ610','PAQ620','PAQ625','PAQ635','PAQ640','PAQ650','PAQ655','PAQ665','PAQ670','PAAQUEX' ])
    tv_comp = ['minutes watch TV or videos past 30 days','# of minutes use computer past 30 days' ]
    recreation = ['minutes vigorous recreational activities','minutes moderate recreational activities']
    for i in tv_comp:
        df[i] = df[i].apply(clean_computer_TV)
    for i in recreation:
        df[i] = df[i].apply(clean_activities)
    for i in df.columns:
        df[i] = df[i].apply(fix_small_numbers)
    for i in tv_comp:
        df[i] = df[i].apply(standardize_time)
    return df
def clean_2013(df):
    to_drop = []
    for i in df.columns:
        if len(i) > 6:
            to_drop.append(i)
    dropped_list = ['PAQ706','PAQ605','PAQ610','PAQ620','PAQ625','PAQ635','PAQ640','PAQ650','PAQ655','PAQ665','PAQ670','PAQ722','PAQ731','PAQ677','PAQ678','PAQ740','PAQ742','PAQ744','PAQ746','PAQ748','PAQ755','PAQ762','PAQ764','PAQ766','PAQ679','PAQ750','PAQ770','PAD615','PAD630','PAD645','PAD680','PAD733']
    for i in dropped_list:
        to_drop.append(i)
    df = df.rename(columns = {'SEQN': 'User ID','PAQ710':'minutes watch TV or videos past 30 days','PAQ715': '# of minutes use computer past 30 days','PAD660':'minutes vigorous recreational activities','PAD675':'minutes moderate recreational activities'}).drop(columns= to_drop)
    tv_comp = ['minutes watch TV or videos past 30 days','# of minutes use computer past 30 days' ]
    recreation = ['minutes vigorous recreational activities','minutes moderate recreational activities']
    for i in tv_comp:
        df[i] = df[i].apply(clean_computer_TV)
    for i in recreation:
        df[i] = df[i].apply(clean_activities)
    for i in df.columns:
        df[i] = df[i].apply(fix_small_numbers)
    for i in tv_comp:
        df[i] = df[i].apply(standardize_time)
    return df
def clean_2015(df):
    to_drop = []
    for i in df.columns:
        if len(i) > 6:
            to_drop.append(i)
    dropped_list = ['PAQ706','PAQ605','PAQ610','PAQ620','PAQ625','PAQ635','PAQ640','PAQ650','PAQ655','PAQ665','PAQ670','PAQ722','PAQ731','PAQ677','PAQ678','PAQ740','PAQ742','PAQ744','PAQ746','PAQ748','PAQ755','PAQ762','PAQ764','PAQ766','PAQ679','PAQ750','PAQ770','PAD615','PAD630','PAD645','PAD680','PAD733']
    for i in dropped_list:
        to_drop.append(i)
    df = df.rename(columns = {'SEQN': 'User ID','PAQ710':'minutes watch TV or videos past 30 days','PAQ715': '# of minutes use computer past 30 days','PAD660':'minutes vigorous recreational activities','PAD675':'minutes moderate recreational activities'}).drop(columns= to_drop)
    tv_comp = ['minutes watch TV or videos past 30 days','# of minutes use computer past 30 days' ]
    recreation = ['minutes vigorous recreational activities','minutes moderate recreational activities']
    for i in tv_comp:
        df[i] = df[i].apply(clean_computer_TV)
    for i in recreation:
        df[i] = df[i].apply(clean_activities)
    for i in df.columns:
        df[i] = df[i].apply(fix_small_numbers)
    for i in tv_comp:
        df[i] = df[i].apply(standardize_time)
    return df
```




We then applied these cleaning functions to each of our year datasets. This would allow for them to have the correct columns so that we could then merge them into one.


```python
## Create overall cleaning functions to drop columns/apply above cleaning functions
## These functions use the above general functions for each dataset (based on year) to clean them for use

df_2007 = clean_2007(df_2007)
df_2009 = clean_2009(df_2009)
df_2011 = clean_2011(df_2011)
df_2013 = clean_2013(df_2013)
df_2015 = clean_2015(df_2015)
```

We then proceeded to merge data from each year starting from 2007 to 2015 into one data file. We merge the data in order to facilitate the process of comparing our variables and develop a better understanding of the potential relationships between them. 


```python
#merge all dfs
df =pd.concat([df_2007, df_2009, df_2011, df_2013, df_2015], sort = True)
```

The data file is based on four columns, two of which represent minutes spent doing moderate or vigorous recreational activity and the other two representing data on the amount of hours people spend on electronics in a 30 day period.
More specifically we are comparing the vigorous recreational activities with the hours of computer usage and the vigorous recreational activities with the hours of TV/video game usage. Next we will compare the moderate recreational activities with computer usage and the moderate recreational activities with the hours of TV/video game usage.

This line of code creates a scatterplot graph comparing the minutes participating in vigorous activities with minutes of daily computer usage.


```python
plt.title('Vigorous recreation activities vs Computer')
plt.ylabel('log minutes spent')
plt.xlabel('index')
plt.scatter(df['minutes vigorous recreational activities'].dropna().index, np.log(df['minutes vigorous recreational activities']).dropna() )
plt.scatter(df['# of minutes use computer past 30 days'].dropna().index, np.log(df['# of minutes use computer past 30 days']).dropna())
plt.legend(['vigorous recreational activities',' computer'])
plt.show()
```

This scatterplot shows that people typically spend more time on computers compared to actively participating in vigorous recreation activities.

```python
#get mean for all columns
def get_mean_of_all(df):
    mean_dict = {}
    for i in df.columns:
        describe = df[i].dropna().describe()
        mean_dict[i] = describe['mean']
    return mean_dict
```

```python
get_mean_of_all(df)
```

```python
df_mean_dict = get_mean_of_all(df)
df_2007_mean_dict = get_mean_of_all(df_2007)
df_2009_mean_dict = get_mean_of_all(df_2009)
df_2011_mean_dict = get_mean_of_all(df_2011)
df_2013_mean_dict = get_mean_of_all(df_2013)
df_2015_mean_dict = get_mean_of_all(df_2015)
```

```python
mean_dict_list = [df_2007_mean_dict,df_2009_mean_dict,df_2011_mean_dict,df_2013_mean_dict,df_2015_mean_dict]
year_ls = [2007,2009,2011,2013,2015]
```

```python
tv_mean_ls = []
comp_mean_ls = []
vig_mean_ls = []
mod_mean_ls = []
for i in mean_dict_list:
    tv_mean_ls.append(i['minutes watch TV or videos past 30 days'])
    comp_mean_ls.append(i['# of minutes use computer past 30 days'])
    vig_mean_ls.append(i['minutes vigorous recreational activities'])
    mod_mean_ls.append(i['minutes moderate recreational activities'])
```

```python
plt.title('Mean TV or Video use over the years')
plt.xlabel('Years')
plt.ylabel('Minutes')
plt.bar(year_ls, tv_mean_ls)
plt.show()
plt.title('Mean computer use over the years')
plt.xlabel('Years')
plt.ylabel('Minutes')
plt.bar(year_ls, comp_mean_ls)
plt.show()
plt.title('Mean vigorous recreation over the years')
plt.xlabel('Years')
plt.ylabel('Minutes')
plt.bar(year_ls, vig_mean_ls)
plt.show()
plt.title('Mean moderate recreation over the years')
plt.xlabel('Years')
plt.ylabel('Minutes')
plt.bar(year_ls, mod_mean_ls)
plt.show()
```
